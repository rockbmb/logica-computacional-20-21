## Aproximação das constantes para o problema

```python
# time é o tempo de travagem máximo que se considera
def constantes_plot(a, b, c, P, time, v_inicial):
    vel_corpo = v_inicial
    vel_rodas = v_inicial
    t = 0
    vels_corpo = [vel_corpo]
    vels_rodas = [vel_rodas]
    ts = [t]
    # discretização do tempo em intervalos de 0.1 segundos.
    dt = 0.1
    x = 0.3
    timer = 0
    m = "free"
    while(t < time and (vel_corpo > 0 or vel_rodas > 0)):
        if timer > x and m == "free":
            c = 4
            m = "stopping"
            timer = 0
        elif timer > x and m =="stopping":
            c = 0.5
            m = "free"
            timer = 0
        timer += dt
        vel_corpo, vel_rodas = vel_corpo + (-c * (vel_corpo - vel_rodas) - b) * dt, vel_rodas + (-a * P + c *(vel_corpo - vel_rodas)) * dt
        t += dt
        vels_corpo.append(vel_corpo)
        vels_rodas.append(vel_rodas)
        ts.append(t)

    fig = plt.figure()
    fig.subplots_adjust(top = 0.8)
    ax1 = fig.add_subplot(211)
    ax1.set_ylabel('velocidade (km/h)')
    ax1.set_xlabel('tempo (s)')
    ax1.set_title('Tempo de travagem com V = v = V_0 = 100 km/h')
    ax1.locator_params(axis='x', integer=True)

    ax1.plot(ts, vels_corpo, ts, vels_rodas)
```

```python
constantes_plot(0.02, 1, 0.5, 2000, 20, 100)

constantes_plot(0.0375, 1, 0.5, 1000, 20, 100)

constantes_plot(0.002, 1.5, 5, 10000, 20, 100)
```
